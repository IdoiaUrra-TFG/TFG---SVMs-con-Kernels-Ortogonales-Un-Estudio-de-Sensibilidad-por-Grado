```python
import numpy as np
import matplotlib.pyplot as plt

# POLINOMIO DE LEGENDRE

def p(x, n):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        p_anteanterior = 1
        p_anterior = x
        for i in range(1, n):
            p_actual = ((2*i + 1) * x * p_anterior - i * p_anteanterior) / (i + 1)
            p_anteanterior = p_anterior
            p_anterior = p_actual
        return p_actual

# PARÁMETROS

n_vals = [1, 2, 5, 10, 20]
x = np.arange(-1, 1, 0.01)

# PANEL 2×3

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
ax_plano = ax.flatten()

for i, n in enumerate(n_vals):
    eje = ax_plano[i]
    eje.plot(x, p(x, n), color='blue')
    eje.set_title(f'n = {n}')
    eje.set_xlabel('x')
    eje.set_ylabel('P(x, n)')

fig.delaxes(ax_plano[5])
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("legendre_polinomios.png", dpi=150, bbox_inches="tight")
plt.show()

# DESCARGA

from google.colab import files
files.download("legendre_polinomios.png")
```
